```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
from IPython.display import display, Markdown

# Ensure plots display in Jupyter Notebook
%matplotlib inline

# Display Project Title
display(Markdown("# **Periodic Table Explorer: A Python-Based Tool for Chemical Element Data Retrieval**"))

# Creating a dataset with 40 elements and their properties
data = {
    "Symbol": ["H", "He", "Li", "Be", "B", "C", "N", "O", "F", "Ne", "Na", "Mg", "Al", "Si", "P", "S", "Cl", "Ar", "K", "Ca",
               "Sc", "Ti", "V", "Cr", "Mn", "Fe", "Co", "Ni", "Cu", "Zn", "Ga", "Ge", "As", "Se", "Br", "Kr", "Rb", "Sr", "Y", "Zr"],
    "AtomicNumber": list(range(1, 41)),
    "IonizationEnergy": [1312, 2372, 520, 899, 801, 1086, 1402, 1314, 1681, 2081,
                         496, 738, 577, 786, 1012, 999, 1251, 1520, 419, 590,
                         633, 659, 650, 652, 717, 762, 758, 737, 745, 906,
                         578, 762, 947, 941, 1140, 1350, 403, 549, 600, 640],
    "Electronegativity": [2.20, None, 0.98, 1.57, 2.04, 2.55, 3.04, 3.44, 3.98, None,
                          0.93, 1.31, 1.61, 1.90, 2.19, 2.58, 3.16, None, 0.82, 1.00,
                          1.36, 1.54, 1.63, 1.66, 1.55, 1.83, 1.88, 1.91, 1.90, 1.65,
                          1.81, 2.01, 2.18, 2.55, 2.96, None, 0.82, 0.95, 1.22, 1.33]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Function to search for an element
def search_element(symbol):
    symbol = symbol.capitalize()
    elem = df[df['Symbol'] == symbol]
    if not elem.empty:
        display(elem)
    else:
        print("Element not found! Please check the symbol and try again.")

# Function to plot ionization energy trend
def plot_ionization_energy():
    plt.figure(figsize=(12, 6))
    plt.bar(df['AtomicNumber'], df['IonizationEnergy'], color='purple')
    plt.title('Ionization Energy Trends Across the Periodic Table')
    plt.xlabel('Atomic Number')
    plt.ylabel('Ionization Energy (kJ/mol)')
    plt.grid(True)
    plt.show()

# Function to plot electronegativity trend (handling None values)
def plot_electronegativity():
    df_filtered = df.dropna(subset=['Electronegativity'])  # Remove NaN values
    plt.figure(figsize=(12, 6))
    plt.plot(df_filtered['AtomicNumber'], df_filtered['Electronegativity'], marker='o', linestyle='-', color='green')
    plt.title('Electronegativity Trends Across the Periodic Table')
    plt.xlabel('Atomic Number')
    plt.ylabel('Electronegativity')
    plt.grid(True)
    plt.show()

# Function to display interactive search
def interactive_search():
    symbol_widget = widgets.Text(value='H', description='Element Symbol:')
    interact(search_element, symbol=symbol_widget)

# Run interactive search in Jupyter Notebook
interactive_search()

# Call functions to display plots
plot_ionization_energy()
plot_electronegativity()
```
